```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.TTF" 
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
df = pd.read_csv('data/01_일별코로나확진자수.csv',encoding = 'euc-KR')
df = df[['확진일','확진자수']]
df.head()
```

```python
# 최근 일주일 현황 파악을 위해 부분 rows 추출

dfw = df.loc[551:]
dfw.reset_index(drop=True, inplace=True)
dfw
```

```python
plt.figure(figsize=(10,9))

# 증가추세를 보기위한 꺾은선그래프(subplot1)과 막대그래프(subplot2)를 시각화 

# 첫번째 꺾은선 그래프
plt.subplot(2,1,1)
plt.plot(dfw['확진일'],dfw['확진자수'], 'o-',ms=3,lw=1.5,label='확진자수')
plt.ylim(0,1000)
plt.xlabel('Date')
plt.ylabel('확진자 수')
plt.legend()

# 두번째 꺾은선 그래프
plt.subplot(2,1,2)
plt.bar(dfw['확진일'],dfw['확진자수'],label='확진자수', width=0.7)
plt.xlabel('Date')
plt.ylabel('확진자 수')
plt.legend()
for i in range(len(dfw['확진일'])):
    plt.text(i,dfw.iloc[i,1], str(dfw.iloc[i,1]) +'명',
            fontsize=9, color='k',horizontalalignment='center',verticalalignment='bottom')

plt.savefig('시각화/01_1_일주일코로나확진자.png', dpi=300)
plt.show()
```

```python
# 최근 한달 현황 파악을 위해 부분 rows 추출

dfm = df.loc[526:]
dfm.reset_index(drop=True, inplace=True)
dfm.head()
```

```python
plt.figure(figsize=(10,9))

ax = plt.subplot(2,1,1)
plt.plot(dfm['확진일'],dfm['확진자수'], 'o-',ms=3,lw=1.5,label='확진자수', color = 'g')
plt.ylim(0,1000)
plt.xlabel('Date')
plt.ylabel('확진자 수')
plt.legend()
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
               
ax = plt.subplot(2,1,2)
plt.bar(dfm['확진일'],dfm['확진자수'],label='확진자수', width=0.7, color = 'g')
plt.xlabel('Date')
plt.ylabel('확진자 수')
plt.legend()
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
for i in range(len(dfm['확진일'])):
    plt.text(i,dfm.iloc[i,1], str(dfm.iloc[i,1]),
            fontsize=9, color='k',horizontalalignment='center',verticalalignment='bottom')

plt.savefig('시각화/01_2_한달코로나확진자.png', dpi=300)
plt.show()
```

# 전국 누적 확진자, 1차 백신 접종자

```python
import pandas as pd

infected = pd.read_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전국 누적확진자.csv', encoding = 'cp949')
infected.set_index('일자',inplace = True)
```

```python
vaccine = pd.read_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전국 백신접종.csv',encoding = 'cp949')
vaccine.sort_values(by = '일자', inplace = True)
vaccine.set_index('일자',inplace = True)
```

```python
infect_vac = pd.concat([infected, vaccine], axis = 1)
infect_vac
```

```python
infect_vac.dropna(inplace = True)
```

```python
infect_vac
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')




ax1 = infect_vac.plot(y = '누적확진자', color = 'Blue', figsize = (15,10))
ax2 = infect_vac.plot(title = '<코로나 일일 누적확진자, 백신 1차 접종자>', fontsize = 13 ,y = ['1차접종','2차접종'], secondary_y = True, ax = ax1, color = ['Orange','Green'], figsize = (15, 10))
ax1.grid(True, axis = 'x')
plt.rc('legend', fontsize=20)

plt.savefig('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/코로나 누적확진자, 백신접종자.png')
```

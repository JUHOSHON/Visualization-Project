```python
import geopandas as gpd
import pandas as pd

# pd.set_option('display.max_rows', None)
map_df = gpd.read_file("C:/Users/SHON/프로젝트연습/TL_SCCO_SIG.shp", encoding = "euc-KR")

map_df.set_index('SIG_KOR_NM', inplace = True)
seoul_df = map_df.loc['종로구':'강동구']
```

```python
seoul_df.drop(['SIG_CD','SIG_ENG_NM'], axis = 'columns', inplace = True)

seoul_df.reset_index(inplace = True)
seoul_df.rename(columns = {'SIG_KOR_NM' : '행정구'}, inplace = True)
seoul_df.set_index('행정구', inplace = True)
seoul_df
```

```python
food_edu = pd.read_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/음식점_학원수.csv', encoding = 'utf-8')
food_edu.set_index('행정구', inplace = True)
food_edu
```

```python
seoul_food =  pd.concat([seoul_df, food_edu['일반음식점수']],
                       axis = 1)
seoul_food.reset_index(inplace = True)
seoul_food
```

```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.TTF" 
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

# 시각화 할 column

variable = '일반음식점수'

# choropleth 범위 설정

vmin, vmax = 0, 15000

# figure, axes 생성

fig, ax = plt.subplots(1, figsize = (15, 9))

seoul_food.plot(column = variable, cmap = 'Greens', linewidth = 0.8, ax = ax, 
           edgecolor = '0.8')
#for i in range(len(merged.index)):
#     plt.annotate(s=merged.index[i], xy=merged.loc[i,'geometry'],
#                horizontalalignment='center')
seoul_food['coords'] = seoul_food['geometry'].apply(lambda x: x.representative_point().coords[:])
seoul_food['coords'] = [coords[0] for coords in seoul_food['coords']]
for idx, row in seoul_food.iterrows():
    plt.annotate(text = row['행정구'], xy=row['coords'],
                horizontalalignment='center', color='black')

ax.axis('off') # 축 제거

# title

ax.set_title('행정구별 일반음식점 분포도', fontdict = {'fontsize':'25','fontweight':'3'})

# 출처 

ax.annotate('Source: data.Seoul. 2021', xy = (0.1, .08),
           xycoords = 'figure fraction', horizontalalignment='left',
           verticalalignment='top',fontsize=12,color='#555555')

# 우측에 범위 표시

sm = plt.cm.ScalarMappable(cmap = "Greens", norm = plt.Normalize(vmin = vmin, vmax = vmax))

sm._A = []
cbar = fig.colorbar(sm)

fig.savefig("C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/음식점_학원수.png", dpi = 300)
```

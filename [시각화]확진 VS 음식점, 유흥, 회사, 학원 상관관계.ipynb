```python
# 일반 음식점 지역별 전처리
import pandas as pd

food = pd.read_csv('C:/Users/SHON/Desktop/KDT\프로젝트/프로젝트 데이터/전처리 후/서울특별시 일반음식점 인허가 정보.csv', encoding = 'cp949')
food = food[food['영업상태']=='영업']
food.dropna(inplace = True)
food.drop('영업상태',axis = 'columns', inplace = True)
food
```

```python
address = food['도로명주소'].str.split(n = 2, expand = True)
address
```

```python
food['행정구'] = address[1]
food.drop('도로명주소', axis = 'columns')
food
```

```python
food = food.groupby('행정구').count()
```

```python
food.rename(columns = {'도로명주소':'일반음식점수'}, inplace = True)
food.to_csv('C:/Users/SHON/Desktop/KDT\프로젝트/프로젝트 데이터/전처리 후/서울시 일반음식점 수.csv')
```

```python
adult_esta = pd.read_csv('C:/Users/SHON/Desktop/KDT\프로젝트/프로젝트 데이터/전처리 후/서울특별시 유흥주점영업 인허가 정보.csv', encoding = 'cp949')
adult_esta
```

```python
adult_esta = adult_esta[adult_esta['영업상태']=='영업/정상']
adult_esta.drop('영업상태',axis = 'columns',inplace = True)
```

```python
address = adult_esta['도로명주소'].str.split(n=2, expand = True)
address
```

```python
adult_esta['행정구'] = address[1]
adult_esta
```

```python
adult_esta = adult_esta.groupby('행정구').count()
adult_esta.rename(columns = {'도로명주소':'유흥주점수'}, inplace = True)
adult_esta
adult_esta.to_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/서울시 유흥주점수.csv')
```

```python
company = pd.read_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/서울시 사업체 현황.csv', encoding = 'cp949')
company.rename(columns = {'자치구':'행정구'}, inplace = True)
company = company[company['동'] == '소계']
```

```python
company.sort_values(by = '행정구', inplace=True)
company.drop('동', axis = 'columns', inplace=True)
company.rename(columns = {'합계':'사업체수'}, inplace=True)
company
#company.to_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/서울시 사업체 수.csv')
```

```python
edu = pd.read_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/서울시 사설학원 통계.csv', encoding = 'cp949')
edu.drop(['합계', '입시검정 및 보습', '국제화', '예능', '특수교육', '종합','기타'], axis = 'columns', inplace = True)
edu.rename(columns = {'소계':'학원수', '자치구':'행정구'}, inplace=True)
#edu.to_csv('C:/Users/SHON/Desktop/KDT/프로젝트/프로젝트 데이터/전처리 후/서울시 학원 수.csv')
```
